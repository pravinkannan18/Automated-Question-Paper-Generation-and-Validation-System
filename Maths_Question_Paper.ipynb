```python
import ollama
import re

# Function to extract text from a text file
def extract_text_from_txt(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
        return text
    except FileNotFoundError:
        print(f"Error: File {file_path} not found.")
        return ""

# Function to process the output stream from Ollama
def output(stream):
    response = ""
    try:
        for chunk in stream:
            content = chunk.get('message', {}).get('content', '')
            print(content, end="")
            response += content
    except Exception as e:
        print(f"\nError during output streaming: {e}")
    return response

# Function to process and extract questions using Ollama
def processone(text):
    prompt = f"""
Extract the questions from the provided text and check for grammatical and spelling mistakes. Correct any mistakes found and provide the clean, corrected text without any unnecessary comments or additional notes. Ensure the text flows naturally and is accurate.

Here is the text:
{text}
"""
    stream = ollama.chat(
        model='llama3.1',
        messages=[{'role': 'user', 'content': prompt}],
        stream=True
    )
    result = output(stream)
    return result

# Function to check irrelevant questions based on syllabus comparison
def check_irrelevant_questions(extracted_questions, syllabus):
    prompt = f"""
You are acting as a syllabus checker. Your task is to carefully review each question and compare it with the syllabus content. Only questions that do not match the syllabus should be marked as irrelevant. 

*Extracted Questions:*
{extracted_questions}

*Syllabus Content:*
{syllabus}

*Instructions:*
- Be strict about comparing each question to the syllabus without assumptions.
- Output only the irrelevant questions, each preceded by its question number (e.g., "Q.No 16.a").
"""
    stream = ollama.chat(
        model='llama3.1',
        messages=[{'role': 'user', 'content': prompt}],
        stream=True
    )
    result = output(stream)
    return result

# Function to generate replacement questions aligned with Bloom's Taxonomy
def generate_replacement_questions(irrelevant_questions, syllabus):
    prompt = f"""
Your task is to rewrite the following irrelevant questions to make them align with the syllabus and adhere to Bloom's Taxonomy. The questions should cover various cognitive skills, from basic recall to higher-order thinking, progressing in difficulty as specified.

*Irrelevant Questions:*
{irrelevant_questions}

*Syllabus Content:*
{syllabus}

*Instructions:*
1. Rewrite each irrelevant question so that it aligns directly with the syllabus content.
2. Ensure that the difficulty progresses following Bloom's Taxonomy:
   - *Part A*: Focus on lower-level skills (Remember, Understand).
   - *Part B*: Focus on mid-level skills (Apply, Analyze).
   - *Part C*: Focus on higher-level skills (Evaluate, Create).
3. Provide only the new replacement questions, each preceded by its original question number (e.g., "Q.No 16.a").
"""
    stream = ollama.chat(
        model='llama3.1',
        messages=[{'role': 'user', 'content': prompt}],
        stream=True
    )
    result = output(stream)
    return result

# Function to parse questions from text using regex
def parse_questions(text):
    """
    Parses questions from the text based on a question numbering pattern.
    Assumes questions are numbered like "Q.No 1.", "Q.No 1.a", etc.
    Returns a list of tuples: (question_number, question_text)
    """
    # Regular expression to match question numbers and their text
    question_pattern = re.compile(r'(Q\.No\s+\d+(\.\w+)?)\s+(.+?)(?=Q\.No\s+\d+(\.\w+)?|$)', re.DOTALL)
    questions = question_pattern.findall(text)
    parsed_questions = []
    for q in questions:
        q_number = q[0].strip()
        q_text = q[2].strip().replace('\n', ' ')
        parsed_questions.append((q_number, q_text))
    return parsed_questions

# Function to replace irrelevant questions with new ones
def replace_questions_automatically(question_paper_path, irrelevant_questions, replacement_questions):
    # Load the original question paper
    with open(question_paper_path, "r", encoding='utf-8') as file:
        question_paper = file.read()

    print("\nOriginal Question Paper:")
    print(question_paper)

    # Parse the question paper into individual questions
    parsed_paper = parse_questions(question_paper)

    # Create a dictionary for replacements
    replacement_dict = {}
    for line in replacement_questions.splitlines():
        match = re.match(r'(Q\.No\s+\d+(\.\w+)?)\s+(.*)', line)
        if match:
            q_number = match.group(1).strip()
            q_text = match.group(3).strip()
            replacement_dict[q_number] = q_text

    # Reconstruct the updated question paper
    updated_paper = ""
    for q_number, q_text in parsed_paper:
        if q_number in replacement_dict:
            # Replace with the new question
            updated_paper += f"{q_number} {replacement_dict[q_number]}\n\n"
        else:
            # Keep the original question
            updated_paper += f"{q_number} {q_text}\n\n"

    print("\nUpdated Question Paper:")
    print(updated_paper)

    # Save the updated question paper to a new file
    with open("updated_model2.txt", "w", encoding='utf-8') as file:
        file.write(updated_paper)

# Main Execution Flow
def main():
    # Load the syllabus and question paper content
    syllabus = extract_text_from_txt('ML1301.txt')
    question_paper_text = extract_text_from_txt('model2.txt')  # Ensure this is the correct path

    if not syllabus or not question_paper_text:
        print("Error: Missing syllabus or question paper content.")
        return

    # Step 1: Extract and clean questions
    extracted_questions = processone(question_paper_text)
    print("\nExtracted and Cleaned Questions:")
    print(extracted_questions)

    # Step 2: Identify irrelevant questions
    irrelevant_questions = check_irrelevant_questions(extracted_questions, syllabus)
    print("\nIdentified Irrelevant Questions:")
    print(irrelevant_questions)

    # Step 3: Generate replacement questions
    replacement_questions = generate_replacement_questions(irrelevant_questions, syllabus)
    print("\nGenerated Replacement Questions:")
    print(replacement_questions)

    # Step 4: Replace irrelevant questions in the question paper
    replace_questions_automatically("model2.txt", irrelevant_questions, replacement_questions)

if __name__ == "__main__":
    main()
```
